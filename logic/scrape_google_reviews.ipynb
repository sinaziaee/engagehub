# get parks in calgary place ids from google api

```python
import requests
import time

# Replace with your actual API key

# Base URL for the Text Search request
base_url = "https://maps.googleapis.com/maps/api/place/textsearch/json"

# Define the search query
params = {
    "query": "parks in Calgary",
    "key": API_KEY
}

all_place_ids = []

while True:
    response = requests.get(base_url, params=params)
    data = response.json()
    
    # Extract place_ids from the current page of results
    for result in data.get("results", []):
        place_id = result.get("place_id")
        if place_id:
            all_place_ids.append(place_id)
    
    # Check if there's a next page of results
    next_page_token = data.get("next_page_token")
    if next_page_token:
        # It may take a few seconds for the next_page_token to become active.
        time.sleep(2)
        params["pagetoken"] = next_page_token
    else:
        break

print("Retrieved place IDs:")
for pid in all_place_ids:
    print(pid)
```
